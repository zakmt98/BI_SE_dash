```python
uspto = []
for i in range(1,8123):
  article ={}
  print(i)
  file = open('/content/drive/My Drive/USPTO_XML/'+'file'+str(i)+".xml",'r')
  soup = BeautifulSoup(file,'lxml')
  try:
    country = soup.find('country').text
    kind = soup.find('kind').text
  except:
    country = 'none'
    kind = 'none'
      
  try:
    prioriri_claim = soup.find('priority-claims')
    date = prioriri_claim.find('date').text
   #print(date)
  except:
    try:
      prioriri_claim = soup.find('priority-claims')
      date = prioriri_claim.find('date').text
    except: 
      date='none'
  try:
    title = soup.find('invention-title').text
    abstract = soup.find('abstract').text
    organisation = soup.find('orgname').text
  except:
    title = 'none'
    abstract = 'none'
    organisation = 'none'
  inventors = []
  for inventor in soup.find_all('inventor'):
    try:
      name = inventor.find('last-name').text +' '+ inventor.find('first-name').text
      address =  inventor.find('country').text +' '+ inventor.find('city').text
    except:
      name = 'none'
      address = 'none'
      invent={'name': name,'address':address}
    inventors.append(invent)
  class_doc = ''
  try:
    for icpr in soup.find('classifications-ipcr'):
      try:
        class_doc = icpr.find('classification-level').text+'-'+icpr.find('class').text\
        +'-'+icpr.find('subclass').text+'-'+'-'+icpr.find('main-group').text+'-'\
        +'-'+icpr.find('subgroup').text+'-'+'-'+icpr.find('symbol-position').text+'-'\
        +'-'+icpr.find('classification-value').text
      except:
        class_doc = 'none'
  except:
      class_doc = 'none'
  article['title']=title
  article['kind']=kind
  article['classifications']=class_doc
  article['country']=country
  article['date']=date
  article['organisation']=organisation
  article['inventors']=inventors
  article['abstract']=abstract
  uspto.append(article)
  print(article )

```

```python
  df = pd.DataFrame(uspto)
```
